```python
# Independent and Dependent Feature Variables

X = df.iloc[:, :-1]
Y = df.iloc[:, -1]

# Train Test Split
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 123)
print("X_train: {}\nX_test : {}\nY_train: {}\nY_test : {}".format(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape))

# Standardizing Data

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)

import pickle
pickle.dump(scaler, open('Scaling.pkl', 'wb')) # Serialised File
```

```python
from sklearn.linear_model import LinearRegression

LR = LinearRegression()
LR.fit(X_train, Y_train)

print(LR.coef_)
print(LR.intercept_)
print(LR.get_params())

Pred = LR.predict(X_test) # Test with Test Data
Pred
```

```python
# Required Plot to Analyse
```

```python
plt.scatter(Y_test, Pred)
```

```python
Residuals = Y_test - Pred
sn.distplot(Residuals)
```

```python
plt.scatter(Pred, Residuals) # Uniform 
```

```python
# Performance Metric

from sklearn.metrics import mean_squared_error, mean_absolute_error

print(mean_squared_error(Y_test, Pred))
print(mean_absolute_error(Y_test, Pred))
print(np.sqrt(mean_absolute_error(Y_test, Pred)))

# R-Square and Adjusted R-Square

# R^2 = 1 - SSR/SST
# Adjusted R^2 = 1 - [(1-R2)*(n-1)/(n-k-1)]

from sklearn.metrics import r2_score
score = r2_score(Y_test, Pred)
print(score)

print(1 - (1-score)*(len(Y_test) - 1)/(len(Y_test) - X_test.shape[1] - 1))

# New Data Prediction

boston.data[0].reshape(1, -1)

LR.predict(boston.data[0].reshape(1, -1))

LR.predict(scaler.transform(boston.data[0].reshape(1, -1)))

# Pickle the Model File 

import pickle
pickle.dump(LR, open('RegModel.pkl', 'wb')) # Serialised File

pickle_model = pickle.load(open('RegModel.pkl', 'rb'))

pickle_model.predict(scaler.transform(boston.data[0].reshape(1, -1)))
```
